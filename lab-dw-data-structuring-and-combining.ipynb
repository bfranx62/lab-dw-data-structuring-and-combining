```python
# Your code goes here

# Import pandas and numpy
import pandas as pd
import numpy as np

# Save given urls
url_1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# Use url objects to create 3 distinct datframes
df_1 = pd.read_csv(url_1)
df_2 = pd.read_csv(url_2)
df_3 = pd.read_csv(url_3)
```

```python
# Viewing all 3 of these shows that they DO all have the same columns, but some are just named differently. 
# So my first step, before merging or joining, will be to standardize column names so a merge or join happens properly. 

# writes a function for cleaning column names for ease and convenience
def clean_col_names(column_names):
    return column_names.str.lower().str.replace(' ','_')

# Clean column names, lower-casing all names and replacing spaces with underscores
df_1.columns = clean_col_names(df_1.columns)
df_2.columns = clean_col_names(df_2.columns)
df_3.columns = clean_col_names(df_3.columns)

# Renames st columns to state
df_1.rename(columns= {'st':'state'}, inplace=True)
df_2.rename(columns= {'st':'state'}, inplace=True)
```

```python
# Standardize Gender Values

# Check value counts in gender column to see which dataframes need fixing
print(df_1['gender'].value_counts()) # df_1 needs it
print(df_2['gender'].value_counts()) # df_2 needs
print(df_3['gender'].value_counts())

# Standardizes values in gender column to M and F
df_1['gender'] = df_1['gender'].str.replace('female','F').str.replace('Femal','F').str.replace('Male','M')
df_2['gender'] = df_2['gender'].str.replace('female','F').str.replace('Male','M')
```

```python
# Standardize State Values

# Check value counts in state column to see which dataframes need fixing
print(df_1['state'].value_counts()) # needs it
print(df_2['state'].value_counts()) # needs it
print(df_3['state'].value_counts())

# Standardizes values in state column to full state names
# .str.replace('fornia','') removes all 'fornia's so as to prevent correction of California to Californiafornia
df_1['state'] = df_1['state'].str.replace('fornia',"").str.replace('Cali','California').str.replace('WA','Washington').str.replace('AZ','Arizona')
df_2['state'] = df_2['state'].str.replace('AZ','Arizona')
```

```python
# Standardize Education Values

# Check value counts in education column to see which dataframes need fixing
print(df_1['education'].value_counts()) # needs it
print(df_2['education'].value_counts()) # needs it
print(df_3['education'].value_counts())

# Changes Bachelors in education column to Bachelor
df_1['education'] = df_1['education'].str.replace('Bachelors','Bachelor')
df_2['education'] = df_2['education'].str.replace('Bachelors','Bachelor')
```

```python
# Standardize Customer Lifetime Values

# Check value counts in customer lifetime value column to see which dataframes need fixing
print(df_1['customer_lifetime_value'].value_counts()) # needs it
print(df_2['customer_lifetime_value'].value_counts()) # needs it
print(df_3['customer_lifetime_value'].value_counts())

# Strips % from values in customer_lifetime_value column
df_1['customer_lifetime_value'] = df_1['customer_lifetime_value'].str.strip('%')
df_2['customer_lifetime_value'] = df_2['customer_lifetime_value'].str.strip('%')
```

```python
# Standardize Vehicle Class values

# Check value counts in vehicle class column to see which dataframes need fixing
print(df_1['vehicle_class'].value_counts()) # needs it
print(df_2['vehicle_class'].value_counts()) # needs it
print(df_3['vehicle_class'].value_counts()) # needs it

# Classifies values Sports Car, Luxury SUV, and Luxury Car as Luxury in vehicle class column
df_1['vehicle_class'] = df_1['vehicle_class'].str.replace('Sports Car','Luxury').str.replace('Luxury SUV','Luxury').str.replace('Luxury Car','Luxury')
df_2['vehicle_class'] = df_2['vehicle_class'].str.replace('Sports Car','Luxury').str.replace('Luxury SUV','Luxury').str.replace('Luxury Car','Luxury')
df_3['vehicle_class'] = df_3['vehicle_class'].str.replace('Sports Car','Luxury').str.replace('Luxury SUV','Luxury').str.replace('Luxury Car','Luxury')
```

```python
# Checks Customer Lifetime Value type
print(df_1['customer_lifetime_value'].dtype) # needs to become float
print(df_2['customer_lifetime_value'].dtype) # needs to become float
print(df_3['customer_lifetime_value'].dtype)

# Converts customer lifetime value from a string to a float
df_1['customer_lifetime_value'] = df_1['customer_lifetime_value'].astype(float)
df_2['customer_lifetime_value'] = df_2['customer_lifetime_value'].astype(float)
```

```python
# Checks Income type
print(df_1['income'].dtype) # needs to become int - after nulls removed
print(df_2['income'].dtype) 
print(df_3['income'].dtype)
```

```python
# Checks monthly premium auto type
print(df_1['monthly_premium_auto'].dtype) # needs to become int - after nulls removed
print(df_2['monthly_premium_auto'].dtype) 
print(df_3['monthly_premium_auto'].dtype)
```

```python
# Checks total claim amount type
print(df_1['total_claim_amount'].dtype) 
print(df_2['total_claim_amount'].dtype) 
print(df_3['total_claim_amount'].dtype)

# All floats, no changes needed
```

```python
# Standardize number of open complaints

# Check Data Types
print(df_1['number_of_open_complaints'].dtype) # object - change after removing nas
print(df_2['number_of_open_complaints'].dtype) # object - change after removing nas
print(df_3['number_of_open_complaints'].dtype) # int

# Cast integer as string
df_3['number_of_open_complaints'] = df_3['number_of_open_complaints'].astype(str)

# Split the values in number of open complaints at every /
split_values_1 = df_1['number_of_open_complaints'].str.split("/")
split_values_2 = df_2['number_of_open_complaints'].str.split("/")

# Save the second value of each in it's own object
sec_value_1 = split_values_1.str[1]
sec_value_2 = split_values_2.str[1]

# Save number of open complaints as only the second values
df_1['number_of_open_complaints'] = sec_value_1
df_2['number_of_open_complaints'] = sec_value_2
```

```python
# Handling Nulls
# Check for Nulls
print(df_1.isnull().any()) # nulls in every column
print(df_2.isnull().any()) # nulls in gender and customer lifetime value columns
print(df_3.isnull().any()) # nulls in no columns

# First step is to eliminate rows that contain only nulls, and then to check the individual columns again. 
df_1 = df_1.dropna(thresh=1)

# Converts income from a float to an int
df_1['income'] = df_1['income'].astype(int)

# Converts monthly premium auto from a float to an int
df_1['income'] = df_1['income'].astype(int)

# Fill nulls in customer lifetime value with the average of that column
df_1.customer_lifetime_value.fillna(df_1.customer_lifetime_value.mean(), inplace=True)
df_2.customer_lifetime_value.fillna(df_2.customer_lifetime_value.mean(), inplace=True)

# Replaces nulls in gender column with whatever value precedes it. 
### I used bfill instead of .dropna(method = "bfill") because I got a warning that using that method will raise an error in future versions of python. 
# I also used bfill because ffill wasn't working, which suggests that it could be the first value of the column that is a NaN.
df_1.gender.bfill(inplace=True)
df_2.gender.bfill(inplace=True)
```

```python
# Now that all dataframes have been cleaned, it may be possible to cleanly concatenate them

# Concatenate dataframes on axis 0
cdf_0 = pd.concat([df_1,df_2,df_3], axis=0)
cdf_0
```

```python
# Handling Duplicates

# Check for Duplicates
print(cdf_0.duplicated().sum())

# Drop Duplicates
cdf_0.drop_duplicates(keep='first', inplace=True)
```
